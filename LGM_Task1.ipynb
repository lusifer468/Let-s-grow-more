```python
from moviepy.editor import ImageSequenceClip
import os

def create_gif(image_folder, gif_name, fps=10):
    """
    Create a GIF from a folder containing images.

    Parameters:
        image_folder (str): Path to the folder containing images.
        gif_name (str): Name of the output GIF file.
        fps (int): Frames per second for the GIF. Default is 10.
    """
    # Get list of image files in the folder
    image_files = [img for img in os.listdir(image_folder) if img.endswith(".jpg") or img.endswith(".png")]

    # Sort the image files based on their names
    image_files.sort()

    # Create a list of image paths
    image_paths = [os.path.join(image_folder, img) for img in image_files]

    # Create GIF from images
    clip = ImageSequenceClip(image_paths, fps=fps)
    clip.write_gif(gif_name, fps=fps)

    print(f"GIF '{gif_name}' created successfully.")

# Example usage
if __name__ == "__main__":
    # Folder containing images
    image_folder = "images"

    # Name of the output GIF file
    gif_name = "output.gif"

    # Create GIF from images in the folder
    create_gif(image_folder, gif_name, fps=15)
```
